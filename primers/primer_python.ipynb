## Load the data

One small table: exports for four countries and fifteen products, in two years. The values are in US dollars.

```python
import os

import pandas as pd

# The table sits next to this notebook if you downloaded the course folder.
# On Google Colab there is no folder, so we read it straight from GitHub.
TRADE_CSV = "trade_extract.csv"
if not os.path.exists(TRADE_CSV):
    TRADE_CSV = "https://raw.githubusercontent.com/shreyasgm/ecu-complexity-labs/main/primers/trade_extract.csv"

trade = pd.read_csv(TRADE_CSV)
trade.head()
```

## How big is it

The shape is rows by columns. Then the column names.

```python
trade.shape
```

```python
trade.columns
```

## One column

Square brackets pull out a single column.

```python
trade["country"]
```

## Keep some rows

A condition inside the brackets keeps the rows where it is true.

```python
morocco = trade[trade["country"] == "Morocco"]
morocco.head()
```

## Two conditions

Morocco in 2023 only. Each condition sits in its own parentheses, joined with `&`.

```python
morocco_2023 = trade[(trade["country"] == "Morocco") & (trade["year"] == 2023)]
morocco_2023
```

## Sort

Largest exports at the top.

```python
morocco_2023.sort_values("export_value", ascending=False).head()
```

## Group and add up

Total exports per country in 2023, over these products.

```python
year_2023 = trade[trade["year"] == 2023]
year_2023.groupby("country")["export_value"].sum()
```

## Make a new column

Each product's share of Morocco's basket. The shares add up to one.

```python
total = morocco_2023["export_value"].sum()
morocco_2023 = morocco_2023.assign(share=morocco_2023["export_value"] / total)
morocco_2023[["product", "share"]].sort_values("share", ascending=False).head()
```

## From a table to a matrix

Countries down the side, products across the top, for one year. Cells with no exports become zero. This grid is the shape the complexity measures are built on.

```python
matrix = year_2023.pivot_table(
    index="country", columns="product_id", values="export_value", fill_value=0
)
matrix.iloc[:, :5]
```

## A quick chart

Morocco's five largest exports in 2023.

```python
import matplotlib.pyplot as plt

top5 = morocco_2023.sort_values("export_value").tail(5)
plt.barh(top5["product"], top5["export_value"])
plt.xlabel("Export value (US dollars)")
plt.tight_layout()
plt.show()
```
