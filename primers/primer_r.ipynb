## Load the data

One small table: exports for four countries and fifteen products, in two years. The values are in US dollars. The tidyverse gives us `read_csv`, the pipe `%>%`, and `ggplot`.

```r
library(tidyverse)

# The table sits next to this notebook if you downloaded the course folder.
# On Google Colab there is no folder, so we read it straight from GitHub.
trade_csv <- "trade_extract.csv"
if (!file.exists(trade_csv)) {
  trade_csv <- "https://raw.githubusercontent.com/shreyasgm/ecu-complexity-labs/main/primers/trade_extract.csv"
}

trade <- read_csv(trade_csv)
head(trade)
```

## How big is it

The dimensions are rows then columns. Then the column names.

```r
dim(trade)
```

```r
names(trade)
```

## One column

The dollar sign pulls out a single column as a vector.

```r
trade$country
```

## Keep some rows

`filter` keeps the rows where a condition is true. The pipe `%>%` sends the table into the next step.

```r
morocco <- trade %>% filter(country == "Morocco")
head(morocco)
```

## Two conditions

Morocco in 2023 only. Extra conditions go in the same `filter`, separated by commas.

```r
morocco_2023 <- trade %>% filter(country == "Morocco", year == 2023)
morocco_2023
```

## Sort

Largest exports at the top.

```r
morocco_2023 %>% arrange(desc(export_value)) %>% head()
```

## Group and add up

Total exports per country in 2023, over these products.

```r
trade %>%
  filter(year == 2023) %>%
  group_by(country) %>%
  summarise(total = sum(export_value))
```

## Make a new column

Each product's share of Morocco's basket. The shares add up to one.

```r
morocco_2023 <- morocco_2023 %>% mutate(share = export_value / sum(export_value))
morocco_2023 %>% select(product, share) %>% arrange(desc(share)) %>% head()
```

## From a table to a matrix

Countries down the side, products across the top, for one year. Cells with no exports become zero. This grid is the shape the complexity measures are built on.

```r
matrix <- trade %>%
  filter(year == 2023) %>%
  pivot_wider(
    id_cols = country,
    names_from = product_id,
    values_from = export_value,
    values_fill = 0
  )
matrix
```

## A quick chart

Morocco's five largest exports in 2023.

```r
top5 <- morocco_2023 %>% arrange(export_value) %>% tail(5)

ggplot(top5, aes(x = export_value, y = reorder(product, export_value))) +
  geom_col() +
  labs(x = "Export value (US dollars)", y = NULL)
```
